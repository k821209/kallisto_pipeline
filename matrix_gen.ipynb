```python
df_v55 = reduce(lambda a,b : pd.merge(a,b,on="target_id",how='outer'),df_list_v55)
df_v11 = reduce(lambda a,b : pd.merge(a,b,on="target_id",how='outer'),df_list_v11)
```

```python
df_v55.columns = ['target_id','0hr_nd','10min_nd','30min_nd','1hr_nd','2hr_nd','24hr_nd','48hr_nd',
                  '0hr_sd','10min_sd','30min_sd','1hr_sd','2hr_sd','24hr_sd','48hr_sd']
df_v11.columns = ['target_id','0hr_nd','10min_nd','30min_nd','1hr_nd','2hr_nd','24hr_nd','48hr_nd',
                  '0hr_sd','10min_sd','30min_sd','1hr_sd','2hr_sd','24hr_sd','48hr_sd']
```

```python
df_v55['genename'] = df_v55['target_id'].apply(lambda x : '.'.join(x.split('.')[0:2]))
df_v55_ix = df_v55.set_index('genename')
```

```python
df_v11['genename'] = df_v11['target_id'].apply(lambda x : '.'.join(x.split('.')[0:2]))
df_v11_ix = df_v11.set_index('genename')
```

```python
genename = 'Cre14.g612633'
fig = plt.figure(figsize=(8,8))
ax1 = fig.add_axes([0.09,0.1,0.2,0.6])
df = df_v11_ix.loc[genename]
D  = df[df.columns[1:]].values
Y  = sch.linkage(D, method='ward')
Z = sch.dendrogram(Y, orientation='right')
ax1.set_xticks([])
ax1.set_yticks([])
ax1.set_axis_off()
zix = Z['leaves']

ax2 = fig.add_axes([0.53,0.1,0.5,0.6])
cax = ax2.pcolor(D[zix])
ax2.set_yticks(np.arange(0.5, len(df.index), 1))
ax2.set_yticklabels(df['target_id'][zix],ha='right')

ax2.set_xticks(np.arange(0.5, len(df.columns[1:]), 1))
ax2.set_xticklabels(df.columns[1:],rotation='vertical')
fig.colorbar(cax, ticks=[0, 10,20,30,40,50,60], orientation='vertical')
plt.savefig(genename+'.hm.png',bbox_inches='tight')
```
